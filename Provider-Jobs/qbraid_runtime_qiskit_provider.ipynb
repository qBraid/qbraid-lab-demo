## Authentication

IBM Quantum authentication requires an API token. Set the `QISKIT_IBM_TOKEN` environment variable or pass it directly to the provider.

Get your token from [IBM Quantum](https://quantum.ibm.com/).

```python
import os

from qbraid.runtime import QiskitRuntimeProvider

# SENTINEL: Requires IBM Quantum API token and service instance CRN.
# Ensure QISKIT_IBM_TOKEN is set in your environment.

token = os.getenv("QISKIT_IBM_TOKEN") or ""
instance = os.getenv("QISKIT_IBM_INSTANCE") or ""  # Set your IBM Quantum instance service CRN

provider = QiskitRuntimeProvider(token=token, instance=instance)
```

```python
# List available IBM Quantum devices
devices = provider.get_devices()
for device in devices:
    print(device.id)
```

## Submit a Job

IBM Quantum devices accept Qiskit `QuantumCircuit` objects as input.

```python
from qiskit import QuantumCircuit

# Create a simple Bell state circuit
circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])

print(circuit)
```

```python
# SENTINEL: Requires active IBM Quantum credentials and device access.
# Submit a job to an IBM backend.

device = provider.get_device("ibm_kingston")
job = device.run(circuit, shots=100)
result = job.result()
print(result.data.measurement_counts)
```

```python
from qbraid.visualization import plot_histogram

counts = result.data.measurement_counts
plot_histogram(counts)
```
