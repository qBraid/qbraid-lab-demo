## Authentication

Amazon Braket uses standard AWS credentials. Set the following environment variables:
- `AWS_ACCESS_KEY_ID`
- `AWS_SECRET_ACCESS_KEY`

Or configure them via `aws configure`. Get your credentials from the [AWS IAM Console](https://console.aws.amazon.com/iam/).

```python
import os

from qbraid.runtime import BraketProvider

# Requires AWS credentials.

aws_access_key_id = os.getenv("AWS_ACCESS_KEY_ID")
aws_secret_access_key = os.getenv("AWS_SECRET_ACCESS_KEY")

provider = BraketProvider(
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key,
)
```

```python
# List available Braket devices
devices = provider.get_devices()
for device in devices:
    print(device.id)
```

## Submit a Job

Braket devices accept Amazon Braket `Circuit` objects as input.

```python
from braket.circuits import Circuit

# Create a simple Bell state circuit
circuit = Circuit()
circuit.h(0)
circuit.cnot(0, 1)
circuit.measure([0, 1])

print(circuit)
```

```python
# Requires active AWS credentials and Braket access.
# Submit a job to the SV1 simulator.

device = provider.get_device("arn:aws:braket:::device/quantum-simulator/amazon/sv1")
job = device.run(circuit, shots=100)
result = job.result()
print(result.data.measurement_counts)
```

```python
from qbraid.visualization import plot_histogram

counts = result.data.measurement_counts
plot_histogram(counts)
```
