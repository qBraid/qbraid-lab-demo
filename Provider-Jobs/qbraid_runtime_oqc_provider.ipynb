## Authentication

OQC authentication requires an API token. Set the `OQC_AUTH_TOKEN` environment variable or pass the token directly to the provider.

Get your token from [OQC Cloud](https://cloud.oqc.app/).

```python
import os
from getpass import getpass

from qbraid.runtime import OQCProvider

# Requires OQC authentication token.
# Ensure OQC_AUTH_TOKEN is set in your environment.

token = os.getenv("OQC_AUTH_TOKEN") or getpass("Enter your OQC auth token: ")

provider = OQCProvider(token=token)
```

```python
# List available OQC devices
devices = provider.get_devices()
for device in devices:
    print(device.id)
```

## Submit a Job

OQC devices accept OpenQASM circuits as input.

```python
# Create a simple Bell state circuit in OpenQASM
program = """
OPENQASM 2;
include "stdgates.inc";
qubit[2] q;
h q[0];
cx q[0], q[1];
"""

print(program)
```

```python
# Requires active OQC credentials and device access.

device = provider.get_device("qpu:uk:2:d865b5a184")
job = device.run(program, shots=100)
result = job.result()
print(result.data.measurement_counts)
```

```python
from qbraid.visualization import plot_histogram

counts = result.data.measurement_counts
plot_histogram(counts)
```
