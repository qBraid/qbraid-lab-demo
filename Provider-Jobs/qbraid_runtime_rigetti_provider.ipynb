## Authentication

Rigetti QCS authentication requires a refresh token. Set the `RIGETTI_REFRESH_TOKEN` environment variable or pass the token directly to the provider.

Get your credentials from [Rigetti QCS](https://qcs.rigetti.com/).

````python
import os

from qbraid.runtime.rigetti import RigettiProvider

# SENTINEL: Requires Rigetti QCS credentials.
# Ensure RIGETTI_REFRESH_TOKEN is set in your environment.

os.environ["RIGETTI_REFRESH_TOKEN"] = '' # set refresh token
provider = RigettiProvider()
````

````python
# List available Rigetti devices
devices = provider.get_devices()
for device in devices:
    print(device.id)
````

## Submit a Job

Rigetti devices accept Quil programs as input.

````python
# Create a simple Bell state circuit in OpenQASM
program = """
OPENQASM 3.0;
bit[2] b;
qubit[2] q;
h q[0];
cx q[0], q[1];
b[0] = measure q[0];
b[1] = measure q[1];
"""

print(program)
````

### quilc Compiler Requirement

Before submitting a job, you must have the **quilc** compiler server running locally. Rigetti uses quilc to compile Quil programs to native gate sets for their QPUs.

Install and start quilc by following the instructions at [github.com/quil-lang/quilc](https://github.com/quil-lang/quilc). For example, using Docker:

```bash
docker run --rm -it -p 5555:5555 rigetti/quilc -R
```

````python
# SENTINEL: Requires active Rigetti QCS credentials and device access.
# Submit a job to a Rigetti device.

device = provider.get_device("Cepheus-1-108Q")
job = device.run(program, shots=100)
result = job.result()
print(result.data.measurement_counts)
````

````python
from qbraid.visualization import plot_histogram

counts = result.data.measurement_counts
plot_histogram(counts)
````
