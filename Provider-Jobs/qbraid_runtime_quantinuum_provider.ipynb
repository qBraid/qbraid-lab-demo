````python
from qbraid.runtime import QuantinuumProvider

provider = QuantinuumProvider()

# List available Quantinuum devices
devices = provider.get_devices()
for device in devices:
    print(device.id)
````

## Submit a Job

Quantinuum devices accept `pytket` circuits as input.

````python
from pytket.circuit import Circuit

# Create a simple Bell state circuit
circuit = Circuit(2)
circuit.H(0)
circuit.CX(0, 1)
circuit.measure_all()

print(circuit)
````

````python
# Requires active Quantinuum credentials and device access.

device = provider.get_device("H1-1E")  # H1-1 Emulator
job = device.run(circuit, shots=100)
result = job.result()
print(result.data.measurement_counts)
````

````python
from qbraid.visualization import plot_histogram

counts = result.data.measurement_counts
plot_histogram(counts)
````
