# Bell Circuit Examples: Universal Program Compatibility

This notebook demonstrates the qBraid Runtime's ability to accept circuits from virtually any major quantum framework and run them on the same device — no manual conversion needed.

Each cell below constructs a Bell state ($|\Phi^+\rangle = \frac{|00\rangle + |11\rangle}{\sqrt{2}}$) in a different framework and submits it to the **qBraid QIR simulator**. The qBraid transpiler handles conversion automatically.

**Frameworks covered:** Amazon Braket · Qiskit · Cirq · PennyLane · PyQuil · PyTKET · OpenQASM 3 · Stim · CUDA-Q · PyQIR · Qibo

```python
import qbraid
qbraid.about()
from qbraid import ConversionGraph

graph = ConversionGraph()
graph.plot()
```

```python
from braket.circuits import Circuit

from qbraid import QbraidProvider

circuit = Circuit()
circuit.h(0)
circuit.cnot(0, 1)
circuit.measure([0, 1])

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```

```python
from qiskit import QuantumCircuit

from qbraid import QbraidProvider

circuit = QuantumCircuit(2, 2)

circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```

```python
from cirq import Circuit, LineQubit, ops

from qbraid import QbraidProvider

q0, q1 = LineQubit.range(2)
circuit = Circuit(ops.H(q0), ops.CNOT(q0, q1), ops.measure(q0, q1))

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```

```python
import pennylane as qml
from pennylane.tape import QuantumTape

from qbraid import QbraidProvider

with QuantumTape() as tape:
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0, 1])
    qml.sample(wires=[0, 1])

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(tape, shots=100)
```

```python
from pyquil import Program
from pyquil.gates import CNOT, MEASURE, H

from qbraid import QbraidProvider

program = Program()

ro = program.declare("ro", "BIT", 2)
program += H(0)
program += CNOT(0, 1)

program += MEASURE(0, ro[0])
program += MEASURE(1, ro[1])

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(program, shots=100)
```

```python
from pytket.circuit import Circuit

from qbraid import QbraidProvider

circuit = Circuit(2)
circuit.H(0)
circuit.CX(0, 1)
circuit.measure_all()

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```

```python
from qbraid import QbraidProvider

qasm = """
OPENQASM 3;
include "stdgates.inc";
qubit[2] q;
bit[2] b;
h q[0];
cx q[0], q[1];
b[0] = measure q[0];
b[1] = measure q[1];
"""

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(qasm, shots=100)
```

```python
from stim import Circuit

from qbraid import QbraidProvider

circuit = Circuit()
circuit.append("H", [0])
circuit.append("CNOT", [0, 1])
circuit.append("M", [0, 1])

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```

```python
try:
    import cudaq
except ImportError:
    print("cudaq is not available on this platform (requires Linux x86_64 + NVIDIA GPU). Skipping.")
else:
    from qbraid import QbraidProvider

    kernel = cudaq.make_kernel()

    qubits = kernel.qalloc(2)

    kernel.h(qubits[0])
    kernel.cx(qubits[0], qubits[1])

    kernel.mz(qubits)

    provider = QbraidProvider()

    device = provider.get_device("qbraid:qbraid:sim:qir-sv")

    job = device.run(kernel, shots=100)
```

```python
from pyqir import BasicQisBuilder, SimpleModule

from qbraid import QbraidProvider

bell = SimpleModule("bell_pair", num_qubits=2, num_results=2)

qis = BasicQisBuilder(bell.builder)

qis.h(bell.qubits[0])
qis.cx(bell.qubits[0], bell.qubits[1])
qis.mz(bell.qubits[0], bell.results[0])
qis.mz(bell.qubits[1], bell.results[1])

bell.builder.ret(None)

module = bell._module
module.verify()

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(module, shots=100)
```

```python
from qibo import Circuit
from qibo.gates import CNOT, H, M

from qbraid import QbraidProvider

circuit = Circuit(2)
circuit.add(H(0))
circuit.add(CNOT(0, 1))
circuit.add(M(0, 1))

provider = QbraidProvider()

device = provider.get_device("qbraid:qbraid:sim:qir-sv")

job = device.run(circuit, shots=100)
```
