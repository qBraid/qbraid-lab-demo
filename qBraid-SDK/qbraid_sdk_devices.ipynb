# qBraid SDK: Devices & Providers Overview

This notebook provides a unified view of all quantum devices accessible through the **qBraid SDK** (>= 0.12.0).

The `QbraidProvider` aggregates devices from multiple hardware vendors into a single interface, allowing you to:
- Browse all available devices across providers
- Check device status and capabilities
- Submit jobs using qBraid credits (no separate provider credentials needed)

For provider-specific examples using native credentials, see the `Provider-Jobs/` directory.

```python
import os

from qbraid import __version__
from qbraid.runtime import QbraidProvider

print(f"qBraid SDK version: {__version__}")
```

## Initialize Provider

The `QbraidProvider` uses your qBraid API key for authentication. On qBraid Lab, this is automatically configured. Otherwise, set the `QBRAID_API_KEY` environment variable.

```python
api_key = os.getenv("QBRAID_API_KEY")

provider = QbraidProvider(api_key=api_key)
```

## List All Available Devices

The table below shows all devices accessible through qBraid, including simulators and QPUs from multiple vendors.

```python
devices = provider.get_devices()

print(f"{'Device ID':<50} {'Status':<12} {'Provider'}")
print("-" * 80)
for device in sorted(devices, key=lambda d: d.id):
    print(f"{device.id:<50} {device.status().name:<12} {device.id.split(':')[0]}")
```

## Device Details

You can inspect individual device properties:

```python
# Example: Get the qBraid QIR simulator
qir_sim = provider.get_device("qbraid:qbraid:sim:qir-sv")

print(f"Device: {qir_sim.id}")
print(f"Status: {qir_sim.status().name}")
print(f"Metadata: ")
for key, value in qir_sim.metadata().items():
    if isinstance(value, dict) and len(value.items()) >= 3:
        print(f"  {key}:")
        for sub_key, sub_value in value.items():
            print(f"    {sub_key}: {sub_value}")
    else:
        print(f"  {key}: {value}")
```
